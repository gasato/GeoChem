```python
fc = FileChooser('.')
fc.title ='<b>Select csv geoChem table</b>'
fc.filter_pattern='*.csv'
fc.reset(path='.', filename= "RNGGeochem.csv")

display(fc)

```

```python
print('Selected dataset to be procesed: ' + fc.selected_filename)
```

```python
# Reading csv with Pandas and create a dataframe

df = pd.read_csv(fc.selected_filename)

# dataframe style

# Selection of items
itemSel = SelectMultiple(
    options=df.columns.values,
    description='Select Elements',
    disabled=False
)

display( itemSel)
```

```python
dfD = df[list(itemSel.value)]

display(dfD)
```

```python
minStats  = []
maxStats  = []
meanStats = []
stdStats  = []
q10       = []
q50       = []
q90       = []

statsLst = [minStats, maxStats, meanStats, stdStats]

for c in itemSel.value:
    #print(c)
    elementSerie = dfD[c].loc[lambda x: (x > 0)]
    minStats.append(elementSerie.min())
    maxStats.append(elementSerie.max())
    meanStats.append(elementSerie.mean())
    stdStats.append(elementSerie.std())
    
    #Nasty patch
    try:
        q10.append(np.quantile(elementSerie, 0.10))
        q50.append(np.quantile(elementSerie, 0.50))
        q90.append(np.quantile(elementSerie, 0.90))
    except:
        q10.append(0)
        q50.append(0)
        q90.append(0)
    
 
d = {'element':list(itemSel.value), 
     'min': minStats, 
     'max': maxStats, 
     'mean': meanStats, 
     'std': stdStats, 
     'Q10%':q10,
     'Q50%':q50,
     'Q90%':q90
    }

dfStats = pd.DataFrame(d).set_index(['element'])

display(dfStats)
```
